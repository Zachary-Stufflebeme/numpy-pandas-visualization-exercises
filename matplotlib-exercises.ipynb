```python
import matplotlib.pyplot as plt
import math
```

```python
x = list(range(100))
y= [ n ** 2 - n + 2 for n in x ]
```

```python
plt.plot(x,y)
plt.annotate('origin', xy = (0,0), xytext = (0,100), arrowprops = {'facecolor':'grey', 'arrowstyle': '->'})
plt.show()
```

```python
plt.figure(figsize = (12,6))

x = range(100)
y = [math.sqrt(n) for n in x]
plt.subplot(221)
plt.plot(x,y)
plt.title('sqrt(x)')

x = range(100)
y = [n ** 3 for n in x]
plt.subplot(222)
plt.plot(x,y)
plt.title('x ^ 3')

x = range(100)
y = [2 ** n for n in x]
plt.subplot(223)
plt.plot(x,y)
plt.title('2 ^ x')

x = range( 2, 100)
y = [ 1 / (n - 1) for n in x]
plt.subplot(224)
plt.plot(x,y)
plt.title('1 / (x-1)')

plt.suptitle('peep the functions')
plt.show()
```

```python
plt.figure(figsize = (12,6))

x = range(100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y, c = 'red' )


x = range(100)
y = [n ** 3 for n in x]

plt.plot(x,y, c = 'orange')


x = range(100)
y = [2 ** n for n in x]

plt.plot(x,y,c = 'yellow')

x = range( 2, 100)
y = [ 1 / (n - 1) for n in x]

plt.plot(x,y,c = 'green')

plt.xlim(0,100)
plt.ylim(0,100)

plt.legend(['$\sqrt{x}$', '$x^{3}$', '$2^{x}$', '1/(x-1)'])
plt.title('peep the functions all in one')
plt.show()
```
